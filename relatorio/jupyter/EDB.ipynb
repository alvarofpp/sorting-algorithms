# Bibliotecas
- **pandas**: manipulação dos dados;
- **bokeh**: exibição dos dados em formato de gráficos.

```python
# Gerais
import pandas as pd

# Bokeh
from bokeh.models import Legend
from bokeh.plotting import figure, output_notebook, show
```

```python
legends = {
    0 : 'Insertion Sort',
    1 : 'Selection Sort',
    2 : 'Bubble Sort',
    3 : 'Shell Sort',
    4 : 'Quicksort',
    5 : 'Merge Sort',
    6 : 'Radix sort (LSD)',
}

colors = {
    0 : 'blue',
    1 : 'green',
    2 : 'orange',
    3 : 'red',
    4 : 'purple',
    5 : 'brown',
    6 : 'black',
}

casos = {
    0 : 'aleatorio',
    1 : 'crescente',
    2 : 'decrescente',
}
```

```python
# Ler arquivo CSV original
data = pd.read_csv('/var/www/aulas/edb/sort/test/tests.csv', sep=';')
# Agrupa os dados pela ordem de algoritmos, número de elementos no vector e cada caso, nessa ordem
# Realiza-se uma média e depois transforma em um dataframe
dt = data.groupby(['algoritmo', 'n_elementos', 'caso'])['tempo'].mean().to_frame()
# Exporta os dados para um novo arquivo csv
dt.to_csv('tests_out.csv', sep=';')
```

```python
dt = pd.read_csv('~/anaconda3/bin/projects/EDB/tests_out.csv', sep=';')
```

```python
for i in range (0, 7):
    for c in range (0, 3):
        dt_csv = dt[dt['algoritmo'] == i]
        dt_csv = dt_csv[dt_csv['caso'] == c]
        
        csv_name = "csvs/{}-{}.csv".format(legends[i], c)
        csv_name = csv_name.replace(' ', '_')
        header = ['n_elementos', 'tempo']
        dt_csv.to_csv(csv_name, sep=',', columns = header, index=False);
        # w['female'] = dt_csv['algoritmo'].map({'female': 1, 'male': 0})
```

```python
title_comum = 'Tempo que os algoritmos demoram para ordenar um arranjo de "n" elementos.'

graph = [
    figure(plot_width=900, plot_height=600,
               title=title_comum, x_axis_label='Qtde de elementos', y_axis_label='Milissegundos'),
    figure(plot_width=900, plot_height=600,
               title=title_comum, x_axis_label='Qtde de elementos', y_axis_label='Milissegundos'),
    figure(plot_width=900, plot_height=600,
               title=title_comum, x_axis_label='Qtde de elementos', y_axis_label='Milissegundos'),
]

legends_a = {}

for i in range(0, 7):
    for c in range (0, 3):
        line_source = dt[dt['algoritmo'] == i]
        line_source = line_source[line_source['caso'] == c]
        
        line = graph[c].line('n_elementos', 'tempo', line_color=colors[i], line_width=2, 
                   source=line_source)
        
        index = "{}{}".format(c, i)
        legends_a[index] = line
        
legend0 = Legend(items=[
    (legends[0], [legends_a["00"]]), 
    (legends[1], [legends_a["01"]]), 
    (legends[2], [legends_a["02"]]), 
    (legends[3], [legends_a["03"]]), 
    (legends[4], [legends_a["04"]]), 
    (legends[5], [legends_a["05"]]), 
    (legends[6], [legends_a["06"]])
], location=(70,0), orientation="horizontal")

legend1 = Legend(items=[
    (legends[0], [legends_a["10"]]), 
    (legends[1], [legends_a["11"]]), 
    (legends[2], [legends_a["12"]]), 
    (legends[3], [legends_a["13"]]), 
    (legends[4], [legends_a["14"]]), 
    (legends[5], [legends_a["15"]]), 
    (legends[6], [legends_a["16"]])
], location=(70,0), orientation="horizontal")

legend2 = Legend(items=[
    (legends[0], [legends_a["20"]]), 
    (legends[1], [legends_a["21"]]), 
    (legends[2], [legends_a["22"]]), 
    (legends[3], [legends_a["23"]]), 
    (legends[4], [legends_a["24"]]), 
    (legends[5], [legends_a["25"]]), 
    (legends[6], [legends_a["26"]])
], location=(70,0), orientation="horizontal")

graph[0].add_layout(legend0, 'below')
graph[1].add_layout(legend1, 'below')
graph[2].add_layout(legend2, 'below')
```

## Gráficos de cada algoritmo

```python
dt = pd.read_csv('~/anaconda3/bin/projects/EDB/csvs/Shell_Sort.csv', sep=',')

graph = figure(plot_width=900, plot_height=600,
               title='Shell Sort', x_axis_label='Qtde de elementos', y_axis_label='Milissegundos')
aleatorio = graph.line('n-elementos', 'aleatorio', line_color='blue', line_width=2, source=dt)
crescente = graph.line('n-elementos', 'crescente', line_color='red', line_width=2, source=dt)
decrescente = graph.line('n-elementos', 'decrescente', line_color='green', line_width=2, source=dt)

legend = Legend(items=[
    ('Aleatório', [aleatorio]), 
    ('Crescente', [crescente]), 
    ('Decrescente', [decrescente])
], location=(70,0), orientation="horizontal")

graph.add_layout(legend, 'below')

output_notebook()
show(graph)
```
